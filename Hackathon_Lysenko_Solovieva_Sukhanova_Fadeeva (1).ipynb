## Часть первая: Содержание CO2
- Найти данные по керну EPICA Dome C содержание CO2 за 800 тыс. лет.
- Построить график CO2 в зависимости от возраста

```python
import numpy as np
import matplotlib.pyplot as plt

ageBP = []
CO2ppm = []
with open(r'C:\Users\yana3\OneDrive\Desktop\python\hackathon\antarctica2015co2composite.txt', 'r', encoding = 'utf-8') as f:
    lines_list = f.readlines()

for i in range(len(lines_list)-1):
    lines = lines_list[i].split('\t')
    ageBP.append(lines[0]) 
    CO2ppm.append(float(lines[1]))

plt.figure(figsize = (16,9))
plt.plot(ageBP, CO2ppm, color = 'red')
plt.title('Зависимость содержания CO2 от возраста керна', fontsize = 20)
plt.xlabel('Возраст керна', fontsize = 17)
plt.ylabel('Содержание CO2 в частицах на миллион', fontsize = 17)
plt.xlim = ([1, 800000])
plt.gca().tick_params(labelsize = 15)
plt.show()
```

## Часть четвертая: Горные керны

В файле Elbrus_raw_data.txt представлены данные содержания основных ионов в ледяном керне полученном летом 2009 г.

- Построить графики глубины и концентрации аммония (NH$_4$$^+$).\n
- Определить в каком году сформировался горизонт на глубине 35.8-35.9 м

В файле Elbrus_annual_summer.txt находятся значения среднего за летний сезон содержания основных ионов (аммоний, нитраты, сульфаты)

- Определить в какой период времени начался активный рост концентрации и предположить возможные причины

```python
kerns = np.loadtxt(r'C:\Users\yana3\OneDrive\Desktop\python\hackathon\Elbrus_raw_data.txt', unpack=True)

depth = kerns[0][1:] # список значений глубин
ammiak = kerns[4][1:] # список значений содержания аммиака
plt.figure(figsize=(20,40))
plt.plot(ammiak, depth)
plt.xlabel('Содержание NH$_4$$^+$', fontsize = 20) # подписываем ось x
plt.ylabel('Глубина, м', fontsize = 20) # подписываем ось y
plt.gca().invert_yaxis() # для удобства переворачиваем ось y, так как это глубина
plt.title('Зависимость содержания NH$_4$$^+$ от глубины', fontsize = 20)
ax.tick_params(labelsize = 20)
```

Для ответа на вопрос о датировке керна на глубине 38,5-39,5 необходимы данные с датировкой содержаний аммония, однако визуальное сравнение показанного выше графика с графиком из файла Elbrus_annual_summer.txt затруднительно. Так как график содержания аммония не линейный, и значения многократно повторяюся всвязи с сезонными колебаниями, датировка глубины по совпадениям значений невозможна. 
    Метры водного эквивалента это искуственная единица измерения, используемая в гляциологии и обозначающая глубину воды, которая может образоваться, если весь снежный покров полностью растаял. Водный эквивалент снега является произведением высоты снежного покрова в метрах на интегрированную по вертикали плотность в килограммах на кубический метр.
    Сезонность колебаний видна примерно до глубины 75 м

```python
kerns_2 = np.loadtxt(r'C:\Users\yana3\OneDrive\Desktop\python\hackathon\Elbrus_annual_summer.txt', unpack=True)

date = kerns_2[0][1:]
NH4 = kerns_2[1][1:]
NO3 = kerns_2[2][1:]
SO4 = kerns_2[3][1:]
plt.plot(NH4, date)
plt.plot(NO3, date)
plt.plot(SO4, date)
plt.xlabel('Содержание NH4, NO3 и SO4[ppb]', fontsize=18) # подписываем ось x
plt.ylabel('Год', fontsize = 18) # подписываем ось y
NH4_mean = sum(NH4[:19])/20 
NO3_mean = sum(NO3[:19])/20
SO4_mean = sum(SO4[:19])/20 # рассчитываем среднее за 20 лет по 3 показателям 
print('За последние 20 лет по сравнению с доиндустриальным временем концентрация аммония увеличилась в ', NH4_mean/NH4[-1], 'раз')
print('За последние 20 лет по сравнению с доиндустриальным временем концентрация нитратов увеличилась в ', NO3_mean/NO3[-1], 'раз')
print('За последние 20 лет по сравнению с доиндустриальным временем концентрация сульфатов увеличилась в ', SO4_mean/SO4[-1], 'раз')
```

Заметный рост концентраций начался после 1950 года, что можно связать с интенсификацией производства, что связано с техническим прогрессом и ростом населения
